# 데이터 전처리

##안쓰는 변수 제거(columns)

```python
# drop 열 제거/ axis = 0 'index'행에 대해 작용 / axis = 1 'columns' 열에 대해 작용
# axis보다 columns로 삭제하면 axis로 지정안해도됨
print("before",train.shape,test.shape)
```

```python
train = train.drop(['Ticket','Cabin'], axis=1)
test = test.drop(['Ticket','Cabin'], axis=1)
```

```python
combine = [train,test]
print("after", train.shape,test.shape)
```

```python
combine
```

```python
# expand = True 이면 df, expand = False 이면 Series
# combine 메서드는 두 pandas 객체를 func함수를 이용하여 결합하는 메서드
for dataset in combine:
    dataset['Title'] = dataset.Name.str.extract(' ([A-Za-z]+)\.', expand=False)
```

```python
train.head()
```

```python
pd.crosstab(train.Title,train.Sex)
```

## 범주화 > 수치화

```python
# 범주화(여자/남자/ 나머지 Rare로 나누기)
for dataset in combine:
    dataset.Title = dataset.Title.replace(['Col','Capt','Countess','Don','Dr','Jonkheer','Lady','Major','Rev','Sir','Rev'],'Rare')
    dataset.Title = dataset.Title.replace(['Mlle'],'Miss')
    dataset.Title = dataset.Title.replace(['Mrs'],'Miss')
    dataset.Title = dataset.Title.replace(['Mme'],'Mrs')
```

```python
train[['Title','Survived']].groupby(['Title'],as_index=False).mean()
```

```python
# Title 변수를 숫자형 변수로 바꿔줌
title_mapping = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}
for dataset in combine:
    dataset['Title'] = dataset['Title'].map(title_mapping)
    dataset['Title'] = dataset['Title'].fillna(0)

train.head()
```

```python
test.info()
```

```python
# 안쓰는 변수 제거

train = train.drop(['PassengerId','Name'],axis=1)
test = test.drop(['Name'],axis=1)
```

```python
combine = [train,test]
print(train.shape,test.shape)
```

```python
# 성별(Sex) 변수를 숫자 범주형 변수로 바꿔줌
Sex_mapping = {'male':1, 'female':2}
for dataset in combine:
    dataset['Sex'] = dataset['Sex'].map(Sex_mapping).astype(int)
```

```python
train.head()
```

```python
grid = sns.FacetGrid(train,row='Pclass',col='Sex')
grid.map(plt.hist,'Age',alpha=0.7)
grid.add_legend()
plt.show()
```

```python
train.isnull().sum()
```

```python
for dataset in combine:
    dataset['Age'].fillna(dataset.groupby("Title")["Age"].transform("median"), inplace=True)
```

```python
for dataset in combine:
    dataset['Agebin'] = pd.cut(dataset['Age'], 5, labels=[0,1,2,3,4])
```

```python
train.head()
```

```python
for dataset in combine:
    dataset.loc[dataset['Age'] <= 15, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 15) & (dataset['Age'] <= 25), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 25) & (dataset['Age'] <= 35), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 35) & (dataset['Age'] <= 45), 'Age'] = 3
    dataset.loc[(dataset['Age'] > 45) & (dataset['Age'] <= 60), 'Age'] = 4
    dataset.loc[dataset['Age'] > 60, 'Age'] = 5
```

```python
# Embarked의 결측치를 최빈값으로 대체 'S'
freq_port = train.Embarked.dropna().mode()[0]
freq_port

train['Embarked'] = train['Embarked'].fillna(freq_port)
```

# 모델링

```python
from sklearn.neighbors import KNeighborsClassifier #KNN
from sklearn.ensemble import RandomForestClassifier #랜덤포레스트
from sklearn.linear_model import LogisticRegression # 로지스틱 회귀
from sklearn.svm import SVC # 서포트 벡터 머신
from sklearn.model_selection import train_test_split
```

```python
train.head()
```

```python
X = train.drop(['Survived'],axis=1).copy().values # 2차원배열
y = train['Survived'].copy().values # 1차원배열
```

```python
X[:5]
```

```python
y[:5]
```

```python
X.shape, y.shape
```

### 랜덤포레스트

```python
# 랜덤포레스트 객체 생성
random = RandomForestClassifier()
# fit 모델 훈련
random.fit(X,y)
```

```python
# 랜덤포레스트 예측값 확인
tr_pred = random.predict(X)
tr_pred
```

```python
# 훈련셋 정확도
random.score(X,y)
```

### knn

```python
# knn 객체 생성
knn = KNeighborsClassifier()
# fit 모델 훈련
knn.fit(X,y)
```

```python
# knn 정확도 확인
knn.score(X,y)
```

### 로지스틱회귀

```python
# logistic 객체 생성
clf = LogisticRegression()
# fit 모델 훈련
clf.fit(X,y)
```

```python
# logistic 훈련 정확도 확인
clf.score(X,y)
```

### 서포트벡터머신

```python
# decision 객체 생성
vector = SVC()
vector.fit(X,y)
vector.score(X, y)
```
