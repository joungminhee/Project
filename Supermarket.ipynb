# 데이터셋준비

```python
data = '/content/sample_data/Stores.csv'
df = pd.read_csv(data)
```

* Store_Area: 야드 광장에 있는 매장입니다.

* Items_Available: 해당 상점에서 사용할 수 있는 물건 의 수입니다.

* DailyCustomerCount: 한 달 동안 평균 매장을 방문한 고객 수입니다.

* Store_Sales: 매장에서 발생한 (US $) 매출입니다.

```python
# 데이터프레임(df)의 특정 컬럼을 지정해주면 해당하는 히스토그램
sns.distplot(df['Store_Sales'])

plt.grid()
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
# corr(상관관계) 그래프
plt.figure(figsize=(12,8))
cor= df.corr()
sns.heatmap(cor, annot = True, cmap='viridis')
plt.show()

# annot=True , 각 cell의 값 표기 유무
# cmap='YlGnBu' , 히트맵 색 설정
```

```python
df.head()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=df, x='Store_id', y='Store_Sales',
            order=df.sort_values('Store_Sales', ascending=False)[:10].Store_id)
plt.title('Top 10 Store_sales')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=df, x='Store_id', y='Daily_Customer_Count',
            order=df.sort_values('Daily_Customer_Count', ascending=False)[:10].Store_id)
plt.title('Top 10 Daily_Customer_Count')
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(data=df, x='Store_id', y='Items_Available',
            order=df.sort_values('Items_Available', ascending=False)[:10].Store_id)
plt.title('Top 10 Items_Available')
plt.show()
```
